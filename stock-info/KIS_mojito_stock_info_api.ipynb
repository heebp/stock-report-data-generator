```python
import os
from dotenv import load_dotenv
import requests
import json

load_dotenv()

KIS_APP_KEY = os.getenv("KIS_APP_KEY")
KIS_APP_SECRET = os.getenv("KIS_APP_SECRET")
KIS_ACC_NO = os.getenv("KIS_ACC_NO")
DB_HOST = os.getenv("DB_HOST")
DB_USERNAME = os.getenv("DB_USERNAME")
DB_PASSWORD = os.getenv("DB_PASSWORD")
DB_SCHEME = os.getenv("DB_SCHEME")
```

```python
import mojito

broker = mojito.KoreaInvestment(
    api_key=KIS_APP_KEY,
    api_secret=KIS_APP_SECRET,
    acc_no=KIS_ACC_NO
)
```

```python
import pandas as pd

# pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

pd.reset_option('display.max_rows', None)
# pd.reset_option('display.max_columns', None)
```

```python
symbols_kospi = broker.fetch_kospi_symbols()
symbols_kospi[symbols_kospi['그룹코드'] == 'ST' ]
```

```python
symbols_kosdaq = broker.fetch_kosdaq_symbols()
symbols_kosdaq[symbols_kosdaq['그룹코드'] == 'ST']
```

```python
filtered_kospi = symbols_kospi[symbols_kospi['그룹코드'] == 'ST']
filtered_kosdaq = symbols_kosdaq[symbols_kosdaq['그룹코드'] == 'ST']

# 두 DataFrame 합치기
combined_symbols = pd.concat([filtered_kospi, filtered_kosdaq])

# 값 치환 (N -> S, Y -> A)
combined_symbols['거래정지'] = combined_symbols['거래정지'].replace({'N': 'S', 'Y': 'A'})
columns_to_keep = ['단축코드', '한글명', '표준코드', '시가총액', '거래정지']

combined_symbols = combined_symbols[columns_to_keep]

combined_symbols

```

```python
duplicated_shortcodes = combined_symbols[combined_symbols['단축코드'].duplicated(keep=False)]

# 중복된 값이 있는지 여부 확인
if not duplicated_shortcodes.empty:
    print("중복된 shortcode가 존재합니다:")
    print(duplicated_shortcodes)
else:
    print("중복된 shortcode가 없습니다.")
```

```python
combined_symbols.to_csv("stock_info.csv", index=True, encoding="utf-8-sig")
```

```python
from sqlalchemy import create_engine, MetaData, Table
# MySQL 데이터베이스 연결 설정
db_username = DB_USERNAME
db_password = DB_PASSWORD
db_host = DB_HOST
db_name = DB_SCHEME

# 데이터프레임 컬럼명과 데이터베이스 컬럼명을 매핑하는 딕셔너리 생성

# SQLAlchemy 엔진 생성
engine = create_engine(f'mysql+pymysql://{db_username}:{db_password}@{db_host}/{db_name}')

# 메타데이터 객체 생성
metadata = MetaData()

table_name = 'STOCK'
table = Table(table_name, metadata, autoload_with=engine)

# 테이블 컬럼명 확인
db_columns = [column.name for column in table.columns]

db_columns
```

```python
column_mapping = {
    '단축코드': 'SHORT_CODE',
    '서브카테고리' : 'SUB_CATEGORY_ID',
    '한글명': 'STOCK_NAME',
    '표준코드': 'STANDARD_CODE',
    '거래소': 'STOCK_EXCHANGE_MARKET',
    '시가총액': 'MARKET_CAPITALIZATION',
    '시가': 'OPEN_PRICE',
    '종가': 'CLOSED_PRICE',
    '고가': 'HIGH_PRICE',
    '저가': 'LOW_PRICE',
    '거래정지': 'STOCK_TRADE_STATUS',
}

# 컬럼명 매핑 적용
combined_symbols.rename(columns=column_mapping, inplace=True)
```

```python
for column in db_columns:
    if column not in combined_symbols.columns:
        combined_symbols[column] = 1
```

```python
dtype_mapping = {
    'SHORT_CODE': 'str',        # MySQL VARCHAR
    'SUB_CATEGORY_ID': 'int64', # MySQL INT
    'STOCK_NAME': 'str',        # MySQL VARCHAR
    'STANDARD_CODE': 'str',     # MySQL VARCHAR
    'STOCK_EXCHANGE_MARKET': 'str',  # MySQL CHAR(1)
    'MARKET_CAPITALIZATION': 'int64',  # MySQL DOUBLE
    'OPEN_PRICE': 'int64',
    'CLOSED_PRICE': 'int64',
    'HIGH_PRICE': 'int64',
    'LOW_PRICE': 'int64',
    'STOCK_TRADE_STATUS': 'str', # MySQL VARCHAR
}

for column, dtype in dtype_mapping.items():
    combined_symbols[column] = combined_symbols[column].astype(dtype)

# DB에 존재하는 컬럼과 일치하는 컬럼만 선택
filtered_df = combined_symbols[db_columns]

# DataFrame의 500번째부터 1000번째 행까지 선택
subset_df = filtered_df.iloc[500:1000]
```

```python
subset_df
```

```python
# combined_symbols DataFrame을 데이터베이스 테이블에 삽입


insert_sql = """
    INSERT INTO STOCK (SHORT_CODE, SUB_CATEGORY_ID, STOCK_NAME, STANDARD_CODE, 
                       STOCK_EXCHANGE_MARKET, MARKET_CAPITALIZATION, OPEN_PRICE, CLOSED_PRICE, 
                       HIGH_PRICE, LOW_PRICE, STOCK_TRADE_STATUS)
    VALUES (%s, %d, %s, %s, %s, %d, %d, %d, %d, %d, %s)
"""

# 각 행의 데이터를 튜플로 변환하여 삽입
values = [tuple(row) for row in subset_df.values]

values
```

```python


try:
    with connection.cursor() as cursor:
        # 여러 개의 레코드를 한 번에 삽입
        cursor.executemany(insert_sql, values)
        
        # 변경 사항을 커밋하여 데이터베이스에 반영
        connection.commit()
        print(f"{cursor.rowcount} rows inserted into STOCK table.")
finally:
    connection.close()
# combined_symbols.iloc[500:502]
```
